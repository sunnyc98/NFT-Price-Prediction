```python
import numpy as np
import pandas as pd
from google.colab import files
from datetime import timedelta
```

```python
def import_csv(service_name):
  try:
    num_of_sales = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/num_of_sales.csv'.format(service_name)).fillna(0)
  except:
    num_of_sales = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/number_of_sales.csv'.format(service_name)).fillna(0)
  average_usd = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/average_usd.csv'.format(service_name)).fillna(0)
  sales_usd = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/sales_usd.csv'.format(service_name)).fillna(0)
  active_market_wallets = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/active_market_wallets.csv'.format(service_name)).fillna(0)
  unique_buyers = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/{}/unique_buyers.csv'.format(service_name)).fillna(0)

  return[num_of_sales,average_usd,sales_usd,active_market_wallets,unique_buyers]

csvs1 = import_csv("art_blocks_art")
csvs2 = import_csv("axie_infinity_game")
csvs3 = import_csv("bored_ape_yacht_club_collection")
csvs4 = import_csv("creepz_game")
csvs5 = import_csv("crypto_punk_collection")
csvs6 = import_csv("decentraland_metaverse")
csvs7 = import_csv("the_sandbox_metaverse")
csvs8 = import_csv("superrare_art")
```

```python
# max_date_df = csvs[0]
# max_date_len = csvs[0].shape[0]
# for i in csvs:
#   if(i.shape[0] > max_date_len):
#     max_date_len = i.shape[0]
#     max_date_df = i

# # def fill_missing_dates(target_df, max_date_df):
# #   idx = pd.date_range(max_date_df['DateTime'].min(), max_date_df['DateTime'].max())
# #   new_target_df = target_df.reindex(idx)
# #   return new_target_df

def fill_missing_dates(target_df):
  # idx = pd.date_range(target_df['DateTime'].min(), target_df['DateTime'].max())
  try:
    target_df['DateTime'] = pd.to_datetime(target_df['DateTime'], format = '%Y-%m-%d %H:%M:%S')
    target_df.set_index("DateTime", inplace=True)
  except:
    pass
  target_df = target_df.resample('D').interpolate().reset_index()
  
  return target_df 
  
def integrate_into_one_csv(csvs):
  filled_csvs =[]
  for i in csvs:
    filled_csvs.append(fill_missing_dates(i))
  
  
  filled_csvs[0] = filled_csvs[0].drop(filled_csvs[0].columns[-1], axis=1)
  for i in range(len(filled_csvs) -1):
    col_name = filled_csvs[i+1].columns[1]
    filled_csvs[0].loc[:,col_name] = filled_csvs[i+1].loc[:,col_name]
  return filled_csvs[0]

integrated_df1 = integrate_into_one_csv(csvs1)
integrated_df2 = integrate_into_one_csv(csvs2)
integrated_df3 = integrate_into_one_csv(csvs3)
integrated_df4 = integrate_into_one_csv(csvs4)
integrated_df5 = integrate_into_one_csv(csvs5)
integrated_df6 = integrate_into_one_csv(csvs6)
integrated_df7 = integrate_into_one_csv(csvs7)
integrated_df8 = integrate_into_one_csv(csvs8)
```

```python
csv_NASDAQ = pd.read_csv('https://raw.githubusercontent.com/qwqw3535/CS370_Machine_Learning/main/original_data_daily/NASDAQ.csv').fillna(0)
try:
    csv_NASDAQ['date'] = pd.to_datetime(csv_NASDAQ['date'], format = '%Y년 %m월 %d일')
    csv_NASDAQ.set_index("date", inplace=True)
except:
  pass
csv_NASDAQ = csv_NASDAQ.sort_values(by=['date'], ascending=True)
# print(csv_NASDAQ.dtypes)

csv_NASDAQ = csv_NASDAQ.resample('D').ffill()
# csv_NASDAQ.head()

#2017/1/1 과 2일이 빠진것을 보충.
prev_two_rows = csv_NASDAQ.iloc[0:2,:].copy()
prev_two_rows.iloc[1,:] = csv_NASDAQ.iloc[0,:]

concat_csv = pd.concat([prev_two_rows, csv_NASDAQ])
old_index = (concat_csv.index).tolist()
old_index[0] = pd.to_datetime(old_index[2]) - pd.Timedelta(days=2)
old_index[1] = pd.to_datetime(old_index[2]) - pd.Timedelta(days=1)
concat_csv.index = old_index
# concat_csv.head()

# 2022/05/14, 15를 보충.
last_two_rows = concat_csv.iloc[-2:,:].copy()
last_two_rows.iloc[0,:] = concat_csv.iloc[-1,:]

concat_csv = pd.concat([ concat_csv, last_two_rows])
old_index = (concat_csv.index).tolist()
old_index[-2] = pd.to_datetime(old_index[-3]) + pd.Timedelta(days=1)
old_index[-1] = pd.to_datetime(old_index[-3]) + pd.Timedelta(days=2)
concat_csv.index = old_index
# concat_csv.head(1961)
```

```python
concat_csv.to_csv('concat_csv.csv', encoding = 'utf-8-sig') 
files.download('concat_csv.csv')
```
